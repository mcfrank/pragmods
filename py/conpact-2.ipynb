# Introduction: the "social object"/"social anxiety" model of conceptual pacts (and lots of other stuff)

We have a standard Bayesian pragmatic recursive setup, with a continuous lexicon.

An *utterance* is a string of *words*. A *lexicon* specifies, for each word, its affinity to each possible object. These affinities are numbers between 0 and 1, and for each word they sum to 1. The literal listener interprets each utterance by multiplying together the affinities of the words in it, and then applying a prior:

$$
P_{L_0}(\text{obj} | \text{utt}) \propto P_{\text{prior}}(\text{obj}) \times \prod_{\text{word} \in \text{utt}} \text{lex}[\text{word}, \text{obj}]
$$

This can be justified by imagining a lower-level literal speaker $S_{-1}$ who produces utterances by just sampling repeatedly from some multinomial that depends on the object at hand. But parametrizing the model this way makes things easier, because it's easier to think about words as having meanings than to think about meanings as having words.

Notation: $S_{i}$ reasons about $L_{i-1}$, who reasons in turn about $S_{i-2}$. Specifically we have:

$$
P_{L_i}(\text{obj} | \text{utt}) \propto P_{\text{prior}}(\text{obj}) \times P_{S_{i-1}}(\text{utt} | \text{obj})
$$

$$
P_{S_i}(\text{utt} | \text{obj}) \propto \exp(\lambda (\log P_{L_{i - 1}}(\text{obj} | \text{utt}) - \text{cost}(\text{utt})))
$$

$$
\text{cost}(utt) = \sum_{\text{word} \in \text{utt}} \text{cost}(\text{word})
$$

Finally, we add lexical uncertainty/learning.

Our actual agents listen using $L_n$, and speak using $S_{n+1}$, for some $n$ (currently $n = 2$, because AFAIK that's the minimal value needed to get all the various qualitative phenomena). Each agent believes that:
0. There is a thing called a *lexicon*, which guides behaviour (the "social object" assumption), and...
1. ...the other agents know what the lexicon is, and...
2. ...the other agents believe that everyone else knows what the lexicon is, but...
3. ...the other agents are wrong to believe this, because in fact I am not sure what the lexicon is; instead I have a prior over it and try to do my best. (While desperately hoping that no-one catches on -- the "social anxiety" assumption.)

The prior is implemented as a set of Dirichlet priors, one per row of the lexicon.

Because the agents are uncertain, they learn from data. There are various possible assumptions one could make about what data is available to learn from. To start with, let's imagine a game that goes like:
1. The speaker is assigned a target object.
2. The speaker produces an utterance.
3. The listener guesses which object was meant. (The speaker observes this guess.)
4. The listener is informed of the correct object.

So in this case, the data available to the speaker is: how the listener interpreted their utterance. This interpretation is actually made by uncertain-$L_n$, but they attribute it to certain-$L_n$, and thus as revealing something about the underlying lexicon.

And the data available to the listener is: what object was meant, and what utterance was produced for it. This utterance was actually produced by uncertain-$S_{n+1}$, but they attribute it to certain-$S_{n-1}$ (note the subscripts!), and thus as revealing something about the underlying lexicon.

An interesting thing to explore in the future will be other sorts of evidence -- for instance, so long as you start with vaguely similar lexicon priors, it's entirely possible in this model for listeners to learn from what utterances are produced, even if they are never told what they meant! And similarly, for speakers to learn from vague feedback from listeners ("got it" versus "huh?"), even if they are never told which meaning the listener actually selected! There's code for this in the implementation, but I'll leave exploring it for later.

The reasoning behind the phrase "social object": this is the assumption that there is a thing called a 'lexicon', which is a fixed and stable and intersubjective thing, but which each individual only glimpses through noisy data. Under the assumption that it *is* a fixed and stable thing which generates all of this noisy data, Bayesian theory kicks in and tells us how to reason about it. So in this sense, each agent treats the lexicon exactly like it's a physical object, like a table or something -- objecthood is also an abstraction that only exists inside a generative model over sensory data. But for physical objects, this is (approximately) normatively correct, because the world actually has some fixed physical configuration and stable physical rules and these actually are what generate our sensory data, and these are the assumptions that make probability theory normative. For a social object like a lexicon, this causal structure is totally wrong; in fact the actual lexicon has no causal power over our sensory data at all, because it doesn't exist. But if everyone around us buys into the illusion, then it may work out okay anyway -- and let us re-use all the inferential mechanisms we developed for dealing with the physical world!

## Implementation

Right now we do inference by importance sampling. This scales well on two important axes: (1) large utterance spaces, and (2) long dialogues. The reason is that we sample a bunch of particles (lexicons) at the beginning, do the expensive pragmatics calculations once for each, and then updating the weights as new dialogue turns occur is very fast. However, it obviously doesn't scale as well with lexicon dimensionality, for the usual reasons why importance sampling sucks in high dimensions.

Some sort of MH or slice sampling would be very very simple to implement; they'd scale better with lexicon dimensionality, but worse with utterance space size and dialogue length. Some sort of hybrid (particle filter with MH steps?) might be best?

The code is structured to make it very easy to implement new sampling schemes. The Domain object encapsulates all the likelihood calculations, and the Dialogue object provides all the useful tools for converting posterior samples into useful inferences. All one would have to do is swap out the FixedSupportImportanceSampler, which is only a few lines long.

# Kicking the tires

Okay, time to see it go. First import the model code, and define some constants:

```python
import conpact2
import numpy as np
SOFTMAX_WEIGHT = 2
# Resolution of the importance sampler
PARTICLES = 1000
# Listener is L2-with-uncertainty, speaker is S3-with-uncertainty
LISTENER_DEPTH = 2
FIGWIDTH = 20
```

And some utility functions for sampling dialogue paths between our modelled speaker and listener, and for displaying information about the results.

```python
def simulate_dialogues(count, turns_per_dialogue, domain, lexicon_prior=None,
                       listener_depth=LISTENER_DEPTH, particle_seed=0, particle_count=PARTICLES,
                       randomize_object_priors=False):
    dialogues = []
    for i in xrange(count):
        r = np.random.RandomState(i)
        dialogue = conpact2.Dialogue(domain, listener_depth, particle_seed, particle_count, alpha=lexicon_prior)
        dialogues.append(dialogue)
        for turn in xrange(turns_per_dialogue):
            if randomize_object_priors:
                log_object_prior = np.log(r.dirichlet([1] * domain.objects))
            else:
                log_object_prior = None
            obj, utt, interp = dialogue.sample_obj_utt_interp(r, log_object_prior)
            # And the speaker observes this as having been understood correctly by the LISTENER
            # And the listener observes this as having been said with this intended meaning by the speaker LISTENER-1
            dialogue.new_data([conpact2.SDataInterp(listener_depth, utt, interp, log_object_prior=log_object_prior)],
                              [conpact2.LDataUttWithMeaning(listener_depth - 1, utt, obj, log_object_prior=log_object_prior)])
    return dialogues
            
def show_lex_posterior(weights_particles, idx1, idx2, labels=True):
    weights, particles = weights_particles
    masses, x_edges, y_edges = np.histogram2d([p[idx1] for p in particles],
                                              [p[idx2] for p in particles],
                                              range=[[0, 1], [0, 1]],
                                              weights=np.exp(weights))
    imshow(masses.T, aspect="equal",
           extent=(x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]),
           origin="lower", vmin=0, cmap="binary", interpolation="nearest")
    if labels:
        xlabel("P(word %s means obj %s | data)" % idx1)
        ylabel("P(word %s means obj %s | data)" % idx2)
    
def show_dialogue(dialogue, idx1=(0, 0), idx2=(1, 0)):
    figure(figsize=(FIGWIDTH, FIGWIDTH * 1.0 / 8))
    def plot_trace(trace, *args, **kwargs):
        x = trace.keys()
        y = trace.values()
        plot(x, y, *args, **kwargs)
    subplot(1, 2, 1)
    plot_trace(dialogue.traces["P-understood"])
    ylim(-0.05, 1.05)
    xlabel("turn")
    ylabel("P(L understands S)")
    #subplot(5, 3, 2)
    #plot_trace(dialogue.traces["symKL"])
    #xlabel("turn")
    #ylabel("symmetrized KL(S's L dist, L's L dist)")
    subplot(1, 2, 2)
    ld_turns = dialogue.traces["length-dist"].keys()
    ld_trace = dialogue.traces["length-dist"].values()
    for i in xrange(1, dialogue.domain.max_utterance_length + 1):
        plot([ld_turn[i] for ld_turn in ld_trace], label=str(i))
    legend()
    ylim(-0.05, 1.05)
    ylabel("P(utt length)")
    xlabel("turn")
    for side in ["speaker", "listener"]:
        key = side + "-single-word-meanings"
        turns = dialogue.traces[key].keys()
        mats = dialogue.traces[key].values()
        figure(figsize=(FIGWIDTH, FIGWIDTH * 1.1 / len(turns)))
        figtext(0.5, 0.1,
                "%s's model of listener's marginal P(obj | utt), for single word utterances"
                % (side,), color="blue", weight="bold", va="bottom")
        for i, (turn, mat) in enumerate(zip(turns, mats)):
            ax = subplot(1, len(turns), 1 + i)
            matshow(mat, cmap="binary", vmin=0, fignum=False)
            clim(0, 1)
            if i == 0:
                ylabel("words")
                xlabel("objects")
            else:
                ax.set_xticks([])
                ax.set_yticks([])
    for side in ["speaker", "listener"]:
        key = side + "-lexicon-posterior"
        turns = dialogue.traces[key].keys()
        dists = dialogue.traces[key].values()
        figure(figsize=(FIGWIDTH, FIGWIDTH * 1.1 / len(turns)))
        figtext(0.5, 0.1, "%s's posterior over lexicons" % (side,), color="blue", weight="bold", va="bottom")
        for i, (turn, dist) in enumerate(zip(turns, dists)):
            subplot(1, len(turns), 1 + i)
            show_lex_posterior(dist, (0, 0), (1, 0), labels=(i == 0))
            
def show_dialogues(dialogues, **kwargs):
    for i, dialogue in enumerate(dialogues):
        figure(figsize=(FIGWIDTH, 0.5))
        # Apparently you have to create an axes or else ipython will drop your figure entirely. Weird.
        axes(); axis("off");
        figtext(0.5, 0, "Sampled dialogue %s" % (i,), fontsize=20, weight="bold")
        show_dialogue(dialogue, **kwargs)
```

Now for our first trick, let's try a simple 2-word 2-object domain, with equal-cost words (the default), a flat prior on lexicons (also the default), and a flat prior on objects (which is, you guessed it, the default).

The resulting graphs are complicated, so let's walk through them.

First, of course, we ran 5 dialogue samples, so there's one set of graphs for each sampled dialogue.

In each set, the first, upper-left graph shows a very important number: the probability that the listener will understand the speaker given their beliefs at each point in time (marginalizing over target objects and utterances). This starts at 0.5, because the prior on lexicons is totally flat, but quickly rises to near 1, which is good.

The upper-right graph shows what proportion of utterances use 1 word, 2 words, etc. Not very interesting in this case.

The next two rows show how the speaker thinks the listener will interpret each possible one-word utterance, and how the listener will actually interpret each possible one-word utterance, at each point in the interchange. We like to see a nice sparse pattern here, and matching patterns between speaker and listener.

Note that in these simulations, the speaker learns from the listener's interpretation of each utterance, while the listener learns from what the speaker actually meant. These can be totally different things -- and at the beginning of the dialogue, they often will be! So we might expect the speaker and listener to actually learn the opposite lexicon sometimes. And in the later dialogues -- 2 and 3, and especially 4 -- we see this happen, where initially one or both agents have opposite beliefs about how words should be interpreted. But since they each then continue to learn from the other's behaviour, this is an unstable state which they quickly recover from.

Finally, the last two rows directly visualize each agent's posterior over lexicons. For this simple 2x2 world, the lexicon space is 2-dimensional, so we can actually draw the posterior distribution directly. (Under the hood this is a histogram on the particle weights.) The way we map the space of lexicons into two dimensions is, the x-axis indicates the literal listener's belief that word 0 refers to object 0, and the y-axis indicates the literal listener's belief that word 1 refers to object 0. (Their probability of referring to object 1 is just one minus the values here.)

These posteriors are shown every 5 turns.

Notice that the initial state is not quite flat -- this is sampling noise in the particle filter. There are 1000 particles in the 1x1 box. Empirically, their location doesn't seem to bias the final outcome much...

Notice also that the posteriors themselves don't always match very well between the speaker and listener -- probably this has to do with the continued influence of whatever random gibberish they used as training data at the beginning of the dialogue -- but also that this doesn't matter much. E.g. in dialogue 2, the listener is convinced that word 0 refers to object 0, and word 1 refers to object 1. The speaker OTOH also believes that word 1 refers to object 1, but thinks that word 0 probably also refers to object 1 -- but, to a lesser extent than word 1. But once pragmatics kick in, these come out pretty much the same.

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=4, softmax_weight=SOFTMAX_WEIGHT)
dialogues = simulate_dialogues(5, 40, domain)
show_dialogues(dialogues)
```

This ambiguity about whether to attribute the particular observed meanings as arising from literal or pragmatic interpretation should, intuitively, only arise when object priors remain constant. So let's simulate changing the object priors around in every trial, with the current prior always being common knowledge.

Does it make a difference? Not really clear:

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=4, softmax_weight=SOFTMAX_WEIGHT)
dialogues = simulate_dialogues(5, 40, domain, randomize_object_priors=True)
show_dialogues(dialogues)
```

### Reduction in conceptual pacts

The same setup we used before to make using multiple words seem worthwhile. But it doesn't really work. I scaled up the number of particles to reduce the chance it's just a sampling issue in this higher-dimensional space. Various fiddling with the softmax weight, depth of recursion, etc., didn't help.

```python
def ambig_2way_prior(domain, high_alpha=1, low_alpha=0.1):
    assert domain.adjectives == domain.objects
    lexicon_prior = low_alpha * np.ones((domain.adjectives, domain.objects))
    for i in xrange(domain.adjectives):
        lexicon_prior[i, i] = high_alpha
        lexicon_prior[(i + 1) % domain.adjectives, i] = high_alpha
    return lexicon_prior

domain = conpact2.MemoizedDomain(adjectives=3, objects=3, word_cost=0.5, max_utterance_length=3, softmax_weight=SOFTMAX_WEIGHT)
lexicon_prior = ambig_2way_prior(domain)
dialogues = simulate_dialogues(3, 20, domain, lexicon_prior=lexicon_prior, particle_count=PARTICLES * 10)
show_dialogues(dialogues)
```

It also doesn't help to try and tweak the prior so that it thinks that really only each word should refer to just one object, it's just not sure which.

# "Horn" implicatures

Our lexical uncertainty is exactly the sort needed to get the Bergen, Goodman, Levy (henceforth BGL) result for "Horn" implicatures. In fact, in a one-shot game, our model is exactly equivalent to theirs (modulo details like level of recursion and how we represent the lexicon). The reason is that while technically the speaker is uncertain at level $S_{n+1}$, and the listener is uncertain at level $L_{n}$, these are equivalent -- the listener says "how would I interpret this on average, if I knew the actual lexicon?", while the speaker says "how will the listener who knows the lexicon interpret this, on average?".

So given a BGL-style Horn implicature domain, our model also produces the BGL result that at turn 0, there is already an asymmetry where word 0 ("cheap") refers more strongly to object 0 ("common") than word 1 ("expensive") does. Intuitively, adding our sort of learning on top of this should help. A properly rational learner would know that the bias to use "cheap" to refer to the common object was a spurious artifact of the other agent's uncertainty. Our agents, though, don't realize that the other agent is uncertain -- they'll tend to use the good utterance/interpretation because of their lexical uncertainty, and then when this works -- because the other agent is doing the same thing -- they'll actually take this as evidence that the good system is the one that they're actually in, and move their beliefs in that direction, setting up a virtuous cycle.

Whether this actually happens seems to depend a lot on the parameters we use and the luck of the draw. With the parameters we've been using so far, and some costs and weights I pulled out of a hat, then... sometimes this works, and sometimes not. Even if we run for a lot of iterations. But notice that when we do converge to a workable system, it's always the "good" equilibrium, not the "bad" one. It's very hard to fully converge to the "bad" equilibrium, because it's, well, bad -- i.e., no-one wants to actually follow what the lexicon says to do, the word costs are pulling against it, so their behaviour is noisier, so they don't converge.

```python
# A trivial version of Leon's common/rare cheap/expensive domain
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=2,
                                 word_cost=[0.5, 1.0], default_object_prior=[0.8, 0.2])
dialogues = simulate_dialogues(5, 40, domain)
show_dialogues(dialogues)
```

If we turn up the softmax parameter, though, then we seem to always converge strongly to the "good" equilibrium:

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=4,
                                 word_cost=[0.5, 1.0], default_object_prior=[0.8, 0.2])
dialogues = simulate_dialogues(5, 20, domain)
show_dialogues(dialogues)
```

Or, if we keep the same softmax parameter, but turn up the number of levels of pragmatic recursion, then likewise, we seem to strongly converge to the "good" equilibrium with high probability. Notice that we then often have massive uncertainty about our lexicon -- all we know is that it's somewhere in the "good half" of the lexicon space -- which makes sense, since the high degree of pragmatic recursion tends to push all of these lexicons together.

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=2,
                                 word_cost=[0.5, 1.0], default_object_prior=[0.8, 0.2])
dialogues = simulate_dialogues(5, 20, domain, listener_depth=8)
show_dialogues(dialogues)
```

# Dogs and daxes

Suppose we have two objects -- object 0 is a dog, and object 1 is novel. And we have two words -- word 0 is "dog", and word 1 is "dax".

This corresponds to a prior that says word 0 almost certainly refers to object 0, but word 1's meaning is totally unknown.

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=SOFTMAX_WEIGHT)
lexicon_prior = [[5, 1], [1, 1]]
listener = conpact2.FixedSupportImportanceSampler(domain, 0, PARTICLES * 10, alpha=lexicon_prior)
show_lex_posterior(listener.weighted_lexicons(), (0, 0), (1, 0))
```

Now someone says: "Look at the dax!"

What do we think this word refers to?

```python
print "Dog: %0.1f%%. Novel object: %0.1f%%." % tuple(100 * np.exp(listener.marginal_dist_n(LISTENER_DEPTH)[1, :]))
```

And what does hearing it do to our posterior?

```python
show_lex_posterior(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1)]).weighted_lexicons(), (0, 0), (1, 0))
```

That's a bit subtle to look at, let's pretend we heard it several times so we can see the pattern:

```python
show_lex_posterior(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1)] * 4).weighted_lexicons(), (0, 0), (1, 0))
```

```python
print "Dog: %0.1f%%. Novel object: %0.1f%%." % tuple(100 * np.exp(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1)] * 4).marginal_dist_n(LISTENER_DEPTH)[1, :]))
```

Weird! The more we hear "dax" in this context, the more we believe it refers to the novel object. BUT ALSO, the more we hear it, the more we believe that the dog is a dax -- and in fact, is a better dax than the novel object is. Just not as good a dax as it is a dog.

The problem is that we believe that the dog and the novel object have equal probabilities of being referred to. So if we keep hearing "dax" over and over, probably some of those times it's being used to refer to the dog.

If we hear "dog" and "dax" an equal number of times, now we are becoming convinced that "dax" means the novel object:

```python
show_lex_posterior(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1), conpact2.LDataUtt(LISTENER_DEPTH - 1, 0)] * 4).weighted_lexicons(), (0, 0), (1, 0))
```

Note, though, that our original conclusion was actually quite reasonable in some situations. Suppose that "dax" meant animal -- then the posterior we got above was totally correct. In real, we can probably make some guess by looking at a novel object about whether it's likely that a single word will refer to both it and dogs. If it's furry and four-footed (but not a dog), then maybe our prior should be that a word likely *will* refer to both it and dogs. In that case, hearing it will indeed make us think that (notice the dirichlet parameters):

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=SOFTMAX_WEIGHT)
lexicon_prior = [[5, 1], [2, 2]]
listener = conpact2.FixedSupportImportanceSampler(domain, 0, PARTICLES * 10, alpha=lexicon_prior)
show_lex_posterior(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1)] * 4).weighted_lexicons(), (0, 0), (1, 0))
```

Alternatively, if we think it's really unlikely that there would be one word that referred to both dogs and this novel object -- it's made out of metal and has sharp edges and so forth -- then simply hearing the word "dax" in this context leaves us *really* convinced that its literal meaning is the novel object (with a bit of residual uncertainty that it might be just another word for dog).

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=SOFTMAX_WEIGHT)
lexicon_prior = [[5, 1], [0.2, 0.2]]
listener = conpact2.FixedSupportImportanceSampler(domain, 0, PARTICLES * 10, alpha=lexicon_prior)
show_lex_posterior(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1)] * 4).weighted_lexicons(), (0, 0), (1, 0))
```

Of course, another problem is that in real life there are *lots* of other words that could be used -- there's not just one novel word in the world. Probably it's not a coincidence that "dax" is the one we heard. What if there are other words whose meaning we don't know -- "dax", "blicket", "furble"?

```python
domain = conpact2.MemoizedDomain(adjectives=4, objects=2, max_utterance_length=1, softmax_weight=SOFTMAX_WEIGHT)
lexicon_prior = [[5, 1], [1, 1], [1, 1], [1, 1]]
listener = conpact2.FixedSupportImportanceSampler(domain, 0, PARTICLES * 10, alpha=lexicon_prior)
show_lex_posterior(listener.weighted_lexicons(), (0, 0), (1, 0))
```

Now when someone says "Look at the dax!", we're not so sure that it refers to the novel object (though that's still the way to bet):

```python
print "Dog: %0.1f%%. Novel object: %0.1f%%." % tuple(100 * np.exp(listener.marginal_dist_n(LISTENER_DEPTH)[1, :]))
```

But, if we hear "dax" several times, we start becoming convinced of two things: (1) "dax" refers to the novel object. (2) "blicket" and "furble" refer to the dog (!).

Note that unlike all the other graphs so far, this is showing the joint distribution of $P(\text{blicket means dog})$ ($x$-axis) and $P(\text{dax means dog})$ ($y$-axis).

```python
show_lex_posterior(listener.with_data([conpact2.LDataUtt(LISTENER_DEPTH - 1, 1)] * 4).weighted_lexicons(), (2, 0), (1, 0))
```

The reason of course is that as we hear "dax" and not "blicket" being used, we suspect that "dax" is a better way to refer to the novel object than "blicket" is (since we're pretty sure that neither of them is the best way to refer to the dog). So this pushes down the probability that blicket literally refers to the novel object. And since meanings have to sum to 1, this pushes up the probability that blicket refers to dog.

Possibly the moral of the story is, there's no reason that meanings should sum to 1. Is this something that makes sense?

# Learning scalar implicature

I live in a world with two equiprobable objects. I keep hearing someone refer to one of them as "some" (or "glasses"), and to the other as "all" (or "hat"). Unlike in the above section, I'm now not just observing the utterances; I'm also observing the meanings. But this time I start out with a totally flat prior:

```python
domain = conpact2.MemoizedDomain(adjectives=2, objects=2, max_utterance_length=1, softmax_weight=SOFTMAX_WEIGHT)
listener = conpact2.FixedSupportImportanceSampler(domain, 0, PARTICLES * 10)
show_lex_posterior(listener.weighted_lexicons(), (0, 0), (1, 0))
```

After I hear the two utterances 5 times each, my posterior has changed:

```python
heard_some_not_all = conpact2.LDataUttWithMeaning(LISTENER_DEPTH - 1, 0, 0)
heard_all_not_some = conpact2.LDataUttWithMeaning(LISTENER_DEPTH - 1, 1, 1)
show_lex_posterior(listener.with_data([heard_some_not_all, heard_all_not_some] * 4).weighted_lexicons(), (0, 0), (1, 0))
```

The best explanation is that "some" means some-but-not-all, and "all" means all-but-not-some. But I have some substantial residual suspicion that "some" might mean some-or-all, or that "all" might mean some-or-all -- but not both!
